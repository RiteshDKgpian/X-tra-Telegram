```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import roc_curve, auc, confusion_matrix
from tensorflow.keras.models import Sequential
from keras.layers import Dense
```

```python
K_values = range(2, 11)
losses = []
accuracies = []
fprs = []
tprs = []
roc_aucs = []
confusion_matrices = []

for K in K_values:
    fold_losses = []
    fold_accuracies = []

    kf = KFold(n_splits=K, shuffle=True)
    for train_index, val_index in kf.split(X):
        X_train, X_val = X.iloc[train_index], X.iloc[val_index]
        y_train, y_val = y[train_index], y[val_index]
        # y_train, y_val = y.iloc[train_index], y.iloc[val_index]

        # Model definition and training
        model = Sequential()
        model.add(Dense(32, activation='relu', input_dim=65))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
        history = model.fit(X_train, y_train
                            , validation_data=(X_val, y_val)
                            , epochs=50, verbose=0)

        # Store loss and accuracy for each epoch
        fold_losses.append(history.history['loss'])
        fold_accuracies.append(history.history['accuracy'])

        # Predict probabilities for ROC curve
        y_pred_prob = model.predict(X_val)
        fpr, tpr, _ = roc_curve(y_val, y_pred_prob, pos_label=2)
        roc_auc = auc(fpr, tpr)
        fprs.append(fpr)
        tprs.append(tpr)
        roc_aucs.append(roc_auc)

        # Confusion matrix
        y_pred = np.round(y_pred_prob)
        confusion_matrices.append(confusion_matrix(y_val, y_pred))

    # Average loss and accuracy across all folds
    losses.append(np.mean(fold_losses, axis=0))
    accuracies.append(np.mean(fold_accuracies, axis=0))
    print('Losses = ', losses)
    print('Accuracy = ', accuracies)


# Plotting the loss versus epochs
plt.figure(figsize=(10, 5))
for i in range(len(K_values)):
    plt.plot(losses[i], label='K={}'.format(K_values[i]))
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss versus Epochs for different K values in K-Fold Cross Validation')
plt.show()

# Plotting the accuracy versus epochs
plt.figure(figsize=(10, 5))
for i in range(len(K_values)):
    plt.plot(accuracies[i], label='K={}'.format(K_values[i]))
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy versus Epochs for different K values in K-Fold Cross Validation')
plt.show()

# Plotting the ROC curve
plt.figure(figsize=(10, 10))
for i in range(len(K_values)):
    plt.plot(fprs[i], tprs[i], label='K={}, AUC={:.2f}'.format(K_values[i], roc_aucs[i]))
plt.plot([0, 1], [0, 1], linestyle='--', color='r', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.show()

# Plotting the confusion matrix
plt.figure(figsize=(10, 5))
for i in range(len(K_values)):
    plt.subplot(3, 3, i+1)
    sns.heatmap(confusion_matrices[i], annot=True, cmap='Blues', fmt='g', cbar=False)
    plt.title('K={}'.format(K_values[i]))
plt.suptitle('Confusion Matrix for different K values in K-Fold Cross Validation')
plt.tight_layout()
plt.show()
```
